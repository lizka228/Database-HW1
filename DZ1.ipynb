Здесть представлен код для анализа таблиц на 1НФ

```python
import pandas as pd
```

Загрузка таблиц

```python
tr = pd.read_excel("customer_and_transaction.xlsx", sheet_name="transaction") # транзакции
cm = pd.read_excel("customer_and_transaction.xlsx", sheet_name="customer") # покупатели
```

```python
tr.duplicated().sum() # проверка на дубликаты
```

```python
cm.duplicated().sum() # проверка на дубликаты
```

```python
cm
```

Создаём 2 новых столбца

```python
cm["number_street"] = cm["address"].apply(lambda x: x.split()[0])
cm["street"] = cm["address"].apply(lambda x: " ".join(x.split()[1:]))
```

```python
cm
```

```python
cm = cm.drop("address", axis=1)
```

```python
tr.columns
```

```python
cm.columns
```

Проверим, что в таблицах customer_id, transaction_id являются первичными ключами.

```python
len(tr) == tr["transaction_id"].value_counts().sum()
```

```python
len(cm) == cm["customer_id"].value_counts().sum()
```

Все id являются уникальными, значит могут быть первичными ключами.

```python
tr
```

```python
pr = tr[["product_id", "brand", "product_class", "product_line", "product_size"]] # таблица с продуктами
```

```python
pr.duplicated().sum() # проверка на дубиликаты
```

```python
pr = pr.drop_duplicates() # удаление дубликатов
```

```python
pr.duplicated().sum()
```

```python
pr.sort_values("product_id",axis=0, ignore_index=True)
```

Здесь от одного и того же product_id зависит множество разных записей, следовательно, product_id не может быть первичным ключом здесь. Анализируя базу данных, видно, что product_id представляет собой просто наименование продукта, например, "велосипед". А следующие атрибуты являются характеристиками. Таким образом можно составить таблицу с продуктами, добавив столбец с первичным ключом либо составной ключ из 5 полей. Добавляем ключевое поле.

```python
pr.insert(0, "id", pr.index)
```

```python
pr
```

То же самое проделаем для таблицы с адресами.

```python
ad = cm[['postcode', 'state', 'country', 'number_street', 'street']] # таблица с адресами
```

```python
ad.sort_values("postcode")
```

```python
ad.duplicated().sum() # проверка на дубиликаты
```

```python
ad.insert(0, "address_id", ad.index)
```

```python
ad
```

Теперь заменим в таблицах с транзакциями и покупателями части с продуктом и адресом соответственно.

```python
tr = pd.merge(tr, pr, on=['product_id', 'brand', 'product_line', 'product_class', 'product_size'])
```

```python
tr
```

```python
cm = pd.merge(cm, ad, on=['postcode', 'state', 'country', 'number_street', 'street'])
```

```python
cm
```

```python
tr = tr.drop(['product_id', 'brand', 'product_line', 'product_class', 'product_size'], axis=1)
cm = cm.drop(['postcode', 'state', 'country', 'number_street', 'street'], axis=1)
```

```python
tr
```

```python
cm.columns
```

Расставляем столбцы как в структуре и сохраняем csv, чтобы загрузить в dbeaver.

```python
tr = tr[['transaction_id', 'id', 'customer_id', 'transaction_date', 'online_order',
       'order_status', 'list_price', 'standard_cost']]
cm = cm[['customer_id', 'first_name', 'last_name', 'gender', 'DOB', 'job_title',
       'job_industry_category', 'wealth_segment', 'deceased_indicator',
       'owns_car', 'address_id', 'property_valuation']]
```

```python
tr.to_csv("transactions.csv", index=False)
cm.to_csv("customers.csv", index=False)
pr.to_csv("products.csv", index=False)
ad.to_csv("addresses.csv", index=False)
```
